```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
from pprint import pprint

# Import API key
from api_keys import g_key
```

```python
file_to_load1 = "starbucksInLACounty.csv"
file_to_load2 = "Income.csv"

# Read Purchasing File and store into Pandas data frame
starbucks_data = pd.read_csv(file_to_load1)
starbucks_data.head(10)
```

```python
new_s_df = starbucks_data.iloc[:,[2,3,4]]
new_s_df
```

```python
gmaps.configure(api_key = g_key)
```

```python
locations = new_s_df[["latitude","longitude"]]

fig = gmaps.figure()
starbucks_layer = gmaps.symbol_layer(locations, 
                                   fill_color = '#87CEFF', stroke_color = '#87CEFF')
fig.add_layer(starbucks_layer)
fig
```

```python
income_data = pd.read_csv(file_to_load2)
income_data.head(10)
```

```python
new_i_df = income_data.iloc[:,[2,3,4,6]]
new_i_df
```

```python
lower_percentile = new_i_df[(new_i_df['Household Income']<new_i_df['Household Income'].quantile(0.25))]
```

```python
lower_percentile
```

```python
locations = lower_percentile[["Latitude","Longitude"]]

fig = gmaps.figure()
city_layer = gmaps.symbol_layer(locations, 
                                   fill_color = '#FF4500', stroke_color = '#FF4500')
fig.add_layer(city_layer)
fig.add_layer(starbucks_layer)
fig
```

```python
high_income = new_i_df[(new_i_df['Household Income']>new_i_df['Household Income'].quantile(0.75))]
```

```python
high_income
```

```python
locations = high_income[["Latitude","Longitude"]]

fig = gmaps.figure()
city_high_layer = gmaps.symbol_layer(locations, 
                                   fill_color = '#C0FF3E', stroke_color = '#C0FF3E')
fig.add_layer(city_high_layer)
fig.add_layer(starbucks_layer)
fig
```

```python
mid_income = new_i_df[(new_i_df['Household Income']>new_i_df['Household Income'].quantile(0.25))&(new_i_df['Household Income']<new_i_df['Household Income'].quantile(0.75))]
```

```python
mid_income
```

```python
locations = mid_income[["Latitude","Longitude"]]

fig = gmaps.figure()
city_mid_layer = gmaps.symbol_layer(locations, 
                                   fill_color = '#FFFF00', stroke_color = '#FFFF00')
fig.add_layer(city_mid_layer)
fig.add_layer(starbucks_layer)
fig.savefig("crime.png")
fig
```
